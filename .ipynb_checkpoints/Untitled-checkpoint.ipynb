```python
data.drop(columns=["Unnamed: 0","flight"],axis=1,inplace=True)
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data["departure_time"]=le.fit_transform(data["departure_time"])
```

```python
data["class"]=le.fit_transform(data["class"])
data["arrival_time"]=le.fit_transform(data["arrival_time"])
```

```python
data
```

```python
data["source_city"].unique()
```

```python
data=pd.get_dummies(data,columns=["source_city","destination_city","airline"],drop_first=True)
```

```python
data.info()
```

```python
columns_to_convert = ["source_city_Chennai", "source_city_Delhi", "source_city_Hyderabad", 
                      "source_city_Kolkata", "source_city_Mumbai", "destination_city_Chennai", 
                      "destination_city_Delhi", "destination_city_Hyderabad", 
                      "destination_city_Kolkata", "destination_city_Mumbai", "airline_Air_India", 
                      "airline_GO_FIRST", "airline_Indigo", "airline_SpiceJet", "airline_Vistara"]

# Convert selected columns to integer type
data[columns_to_convert] = data[columns_to_convert].astype(int)
```

```python
data.info()
```

```python
data["stops"].unique()
```

```python
data["stops"]=data["stops"].map({"zero":0,"one":1,"two_or_more":2})
```

```python
data.head(2)
```

```python
x=data.drop(columns=["price"])
y=data[["price"]]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
print(xtrain.shape,ytrain.shape)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
xtrain=ss.fit_transform(xtrain)
xtest=ss.transform(xtest)
```

```python
from sklearn.linear_model import LinearRegression
lg=LinearRegression()
lg.fit(xtrain,ytrain)
y_pred=lg.predict(xtest)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mean_squared_error(ytest,y_pred)
mean_absolute_error(ytest,y_pred)
r2_score(ytest,y_pred)   
```
